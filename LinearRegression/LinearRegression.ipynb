```python
x = data.drop('Price' , axis=1)
x
```

```python
y = data.Price
```

```python
# data 100% -> train 80%  , test  20%

# shuffle -> mix data
# test_size -> 0.2 -> test -> 20% , train ->80%
# random_state -> =42  ttake the same samples

from sklearn.model_selection import train_test_split

X_train, X_test , y_train   , y_test = train_test_split(x , y , shuffle=True , test_size=0.2 , random_state=42)
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()

X_train[['Avg. Area Income' , 'Area Population']] = scalar.fit_transform(X_train[['Avg. Area Income' , 'Area Population']] )
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()

X_test[['Avg. Area Income' , 'Area Population']] = scalar.fit_transform(X_test[['Avg. Area Income' , 'Area Population']] )
```

```python
# model 

from sklearn.linear_model import LinearRegression
```

```python
# X_train , Y_train 
# model
# training

model = LinearRegression()


```

```python
model.fit(X_train , y_train)
```

## PREDICTION

```python
## Preduict

y_pred = model.predict(X_test)
```

```python
y_pred
```

## model evaluation

```python
from sklearn.metrics import mean_squared_error  , mean_absolute_error
```

```python
mean_squared_error(y_test , y_pred)
```
